```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
file = open("Source_data_for_CFR_vaccine_map_abridged.csv", 'r')

data_s = []

for line in file:
    source = line.strip("\n").split(',')
    data_s.append(source)

data = np.array(data_s)
print(data)
```

```python
fig = plt.figure(figsize=(14,10))
ax = plt.subplot(1,1,1)

earth = Basemap(projection='mill')

earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

plt.title("Measles Outbreak - 100555398", fontsize =18)

#for each 'Outbreak', a for loop goes through the data to make arrays 
#specfic to the outbreak. Each outbreak array is put into futher sub-arrays 
#used to plot coordinates and radii of each point

measles = []

for row in data:
    for element in row:
        if element == 'Measles':
            measles.append(row)

x = []
for row in measles:
        x.append(row[1])

y = []
for row in measles:
        y.append(row[2])
        
radius = []
for row in measles:
    radius.append(row[3])
s = max(radius)

plt.scatter(x,y, color = 'red')
plt.show()
```
