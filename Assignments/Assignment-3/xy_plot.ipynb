```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = open("xy_points.dat" , 'r')

x = np.empty(1000)
y = np.empty(1000)

line = data.readline().split(' ')
for i in range(0, len(line)):
    x[i] = line[i]

line = data.readline().split(' ')
for i in range(0, len(line)):
    y[i] = line[i]

    
x_c = x[ (x < -2) ]
y_c = y[ (x < -2) ]
x_partc = x_c[ (y_c > -5) & (y_c < 0) ]
y_partc = y_c[(y_c > -5) & (y_c < 0)]
x_partd = x[((((x)**2 + (y-5)**2)**(1/2)) <= 5)]
y_partd = y[((((x)**2 + (y-5)**2)**(1/2)) <= 5)]



plt.scatter(x, y, color= 'red')
plt.scatter(x_partc, y_partc, color= 'blue')
plt.scatter(x_partd, y_partd, color= 'green')

plt.title("Highlighted points (100555398)", fontsize =14)
plt.axis([-10, 10, -10, 10])
plt.show()

```
