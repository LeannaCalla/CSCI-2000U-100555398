```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches #for my ledgend
import calendar
import os

files = os.listdir()
temperature_files = []
    
for file in files:
    if 'temperature_2012.dat' in file:   
        temperature_files.append(file)
print(temperature_files)
#checks to see that correct files are being printed

plt.plotfile(temperature_files[0], color = 'purple', linewidth='2', names=calendar.month_abbr[1:13])
plt.plotfile(temperature_files[1], color = 'red', linestyle = '-.', linewidth= '2', newfig= False, names=calendar.month_abbr[1:13])
plt.plotfile(temperature_files[2], color = 'blue', linestyle= '--', linewidth= '2', newfig = False, names=calendar.month_abbr[1:13])

plt.title("2012 average monthly temeratures (100555398)", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Average Monthly temperature (C)")

purple_patch = mpatches.Patch(color='purple', label='Toronto')
red_patch = mpatches.Patch(color='red', label= 'Vancouver')
blue_patch = mpatches.Patch(color='blue', label= 'Montreal')
plt.legend(handles=[purple_patch, red_patch, blue_patch])

plt.xticks(np.arange(12), calendar.month_abbr[1:13])
plt.show()
```
